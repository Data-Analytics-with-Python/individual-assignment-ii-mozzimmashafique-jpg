### Task A.1 (15 points)

```python
import numpy as np
import plotly.express as px

# Create a single campaign response column
df['AcceptedAny'] = np.where(
    df[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']].sum(axis=1) > 0,
    'Yes',
    'No'
)

# Task A.1 scatter plot
fig_a1 = px.scatter(
    df,
    x='NumWebVisitsMonth',
    y='NumWebPurchases',
    size='Income',
    color='AcceptedAny',
    hover_data={
        'Income': True,
        'Education': True,
        'Marital_Status': True
    },
    size_max=35,
    title='Task A.1 — Website Visits vs Online Purchases'
)

fig_a1.show()




```

**Explanation**

**Do more frequent site visitors tend to buy more?**

From the chart, customers who visit the website more frequently generally tend to make more online purchases. The points become denser at higher purchase values as the number of visits increases, especially between 3 to 7 monthly visits.

**How does income affect online purchases?**

Income, shown by bubble size, does not show a strong or consistent relationship with online purchases. Both higher-income and lower-income customers appear across various purchase levels, indicating that income alone does not determine how much customers buy online.

**Are campaign responders visually distinguishable? If so, how are they different?**

Campaign responders are visually distinguishable through the color coding. Although fewer in number, those who accepted a campaign tend to appear more often in the mid-to-higher purchase range (around 6 to 11 purchases), suggesting they may be more engaged and responsive customers.

## Part B (30 points)

In this part, we will study customers' pattern/distribution of spendings across different types of customers. We will visualize such patterns for different customer groups using **bar charts**.  

**Hint**: In this exercise, you may need to use the `df.melt()` function to convert a wide table into a long table. You can find an example of this function in the assignment repository (see **melt_example.ipynb**).

### Task B.1 (15 points)

i. Consider that customers are grouped by their **education levels**. Develop a bar chart with several sets of bars, in which:

* Each set of bars (i.e.,bar of the same color) represent the spendings by customers in different groups on a spefic product category. For example, spending by customers "Graduation", "Master", "PhD", ..., on *fish products*.  
* The chart should allow you to easily visuallize both the **total spending** by customers in a group on all product categories and the **composition** of their spending on those product categories.

ii. Further, answer the following questions in a **text (markdown) cell** and make sure to explain your answer based on the chart:
* Which customer group contributes the highest revenue?

**Hint**: By default, each row in the dataset is shown as a separate rectangle in the bar, and these rectangles have borders that can make the chart look cluttered. In that case, use `fig.update_traces(opacity=1, marker_line_width=0)` to remove the border lines.

```python
import plotly.express as px

# Melt spending columns into long format
spend_cols = ['MntWines','MntFruits','MntMeatProducts',
              'MntFishProducts','MntSweetProducts','MntGoldProds']

long_df = df.melt(
    id_vars='Education',
    value_vars=spend_cols,
    var_name='Category',
    value_name='Spend'
)

# Aggregate spend by Education + Category
edu_agg = long_df.groupby(['Education','Category'], as_index=False)['Spend'].sum()

# Plot stacked bar chart
fig_b1 = px.bar(
    edu_agg,
    x='Education',
    y='Spend',
    color='Category',
    title='Task B.1 — Spending by Education Group (Stacked by Category)'
)

fig_b1.update_traces(opacity=1, marker_line_width=0)  # remove borders
fig_b1.show()
```

**Explanation**

**Which customer group contributes the highest revenue?**

From the stacked bar chart, the Graduation group contributes the highest total revenue. Their bar is significantly taller than all other education groups, showing the largest combined spending across wines, meat products, gold products, and other categories. This indicates that customers with a Graduation-level education are the highest-spending segment overall.

```python
import plotly.express as px

# Melt spending columns into long format
spend_cols = ['MntWines','MntFruits','MntMeatProducts',
              'MntFishProducts','MntSweetProducts','MntGoldProds']

long_m = df.melt(
    id_vars='Marital_Status',
    value_vars=spend_cols,
    var_name='Category',
    value_name='Spend'
)

# Aggregate spend by Marital Status + Category
marital_agg = long_m.groupby(['Marital_Status','Category'], as_index=False)['Spend'].sum()

# Plot grouped bar chart
fig_b2 = px.bar(
    marital_agg,
    x='Marital_Status',
    y='Spend',
    color='Category',
    barmode='group',
    title='Task B.2 — Spending by Marital Status (Grouped by Category)'
)

fig_b2.update_traces(opacity=1, marker_line_width=0)
fig_b2.show()
```

**Explanation**

**For each marital status group, which category of product do they spend the highest amount on?**

Based on the grouped bar chart, Wines (MntWines) is the highest-spending category across almost all marital status groups.
Here is the breakdown:

Married: Highest spending on Wines, followed by Meat Products.

Together: Highest spending on Wines, then Meat Products.

Single: Highest spending on Wines, then Meat Products.

Divorced: Highest spending on Wines, followed by Meat Products.

Widow: Highest spending on Wines.

Alone, Absurd, YOLO Spending levels are very low, but Wines is still slightly higher than other categories.

Across all groups, Wines consistently dominates as the top spending category, indicating strong overall preference.

**Does the marital status affect customers' spending distribution across the product categories? How?**

Yes, marital status does appear to influence spending patterns:

Married, Together, and Single customers show the highest overall spending, with much larger bars in all product categories.

These three groups also exhibit a more balanced spending mix, where Wines and Meat Products account for the largest share but other categories are also noticeably present.

Divorced customers spend moderately, with a similar preference pattern but in smaller amounts.

Widowed, Alone, Absurd, and YOLO groups have minimal spending, showing that they contribute very little to revenue, and their distribution is heavily skewed toward Wines only.

Overall, marital status impacts both total spending level and spending composition, with partnered or formerly partnered groups (Married, Together, Divorced) spending significantly more and across more categories.

## Part C (30 points)

In this part, let us investigate variability in customers' online purchasing behavior.

i. We will use a visualization chart that shows distribution of customers' online purchases. The chart should:
* Show multiple distributions, each being the distribution of number of online purchases by customers of a certain generation (Gen X vs. Millennials,  ignore other generations)
* Allow users to observe the 1st, 2nd, and 3rd quartiles of the distribution, along with any outliers.

ii. Based on the chart, answer the following question:
* How would you comment on the pattern of online purchases across different generations.

iii. In your chart, further divide each distribution into subgroups of customers based on the **number of kids at home** (not including teens).
How would you comment on the pattern of online purchases across the different subgroups of customers?




```python
import plotly.express as px

# Create Generation column (based on common definitions)
df['Generation'] = df['Year_Birth'].apply(
    lambda x: 'Millennial' if 1981 <= x <= 1996
    else ('Gen X' if 1965 <= x <= 1980 else 'Other')
)

# Filter only Gen X and Millennials
df_gen = df[df['Generation'].isin(['Gen X', 'Millennial'])]

# Box plot showing quartiles + outliers
fig_c = px.box(
    df_gen,
    x='Generation',
    y='NumWebPurchases',
    points='outliers',
    title='Part C — Online Purchase Distribution (Gen X vs Millennials)'
)

fig_c.show()
```

**Explanation**

**How would you comment on the pattern of online purchases across different generations.**

From the chart, there are clear differences in online purchasing behavior depending on how many kids a customer has at home:

• Customers with no kids tend to make the highest number of online purchases.
They show a wider distribution and higher median values, suggesting they shop online more frequently and more consistently.

• Customers with 1 kid show moderate online purchase activity.
Their distribution is slightly lower than those with no kids, but still reasonably active.

• Customers with 2 or more kids generally make fewer online purchases.
Their quartiles and medians are noticeably lower, and their spread is tighter, indicating both:

less frequent online purchasing

fewer high-purchase outliers

Overall, the pattern suggests that as the number of kids increases, online purchases tend to decrease.
Customers with more children likely face higher time and financial constraints, reducing their online shopping frequency.

```python
import plotly.express as px

# Create Generation column
df['Generation'] = df['Year_Birth'].apply(
    lambda x: 'Millennial' if 1981 <= x <= 1996
    else ('Gen X' if 1965 <= x <= 1980 else 'Other')
)

# Filter only Gen X and Millennials
df_gen = df[df['Generation'].isin(['Gen X', 'Millennial'])]

# Box plot split by number of kids at home
fig_c3 = px.box(
    df_gen,
    x='Generation',
    y='NumWebPurchases',
    color='Kidhome',
    points='outliers',
    title='Part C — Online Purchases by Generation and Number of Kids at Home',
    labels={'Kidhome': 'Kids at Home'}
)

fig_c3.show()
```

**Explanation**

**How would you comment on the pattern of online purchases across the different subgroups of customers?**

Based on the boxplot comparing online purchases by generation and number of kids at home, several patterns appear:

• Customers with no kids consistently make the highest number of online purchases across both Gen X and Millennials.
Their medians are higher, and they show more upper-end outliers, meaning they purchase online more frequently and with greater variability.

• Customers with 1 kid make fewer online purchases than those with no kids.
Their distributions are tighter, with lower medians and fewer extreme values.

• Customers with 2 kids generally make the fewest online purchases.
Both generations show the lowest medians and the smallest spread for this group, indicating less frequent and more uniform purchasing behavior.

Overall, online purchasing decreases as the number of kids at home increases.
This trend is true for both Gen X and Millennials, suggesting that having more children may limit time or discretionary spending available for online shopping.
